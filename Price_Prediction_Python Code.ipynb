# Correlation identification between price and numerical input features using Bar graph and a Heatmap

```python
numerical_features = ['symboling', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type',
                      'engine-size', 'fuel-system', 'bore', 'stroke',
                      'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg',
                      'highway-mpg']

# Calculate the correlation between each features and "price"
correlation = df[numerical_features].corrwith(df['price'], numeric_only=True)

# Plot the correlation values
plt.figure(figsize = (10, 8))

# orientation = horizontal
sns.barplot(x = correlation.values, y = correlation.index, orient = 'h')

plt.xlabel('Correlation')
plt.ylabel('Features')
plt.title('Correlation of Input Features with Price')
plt.show()
```

```python
sns.set(rc = {'figure.figsize':(12,8)})
dataplot = sns.heatmap(df.corr(),annot=True)
```

```python
from scipy.stats import chi2_contingency

categorical_columns = df.select_dtypes(include='object')

chi2_results = {}

# Iterate over each feature and calculate the chi-square statistic and p-value
for feature in categorical_columns:

    contingency_table = pd.crosstab(df[feature], df['price'])

    # performs chi square test
    chi2, p_value, _, _ = chi2_contingency(contingency_table)

    # Store the chi-square statistic and p-value into a dictionary
    chi2_results[feature] = {'chi2_statistic': chi2, 'p_value': p_value}


tempArr = [] # tempArr to store features that are < 0.05
for feature, result in chi2_results.items():
    print(f"Chi-square test results for feature '{feature}':")
    print(f"Chi-square statistic: {result['chi2_statistic']}")
    print(f"P-value: {result['p_value']}")
    print()
```

# To check for nominal data's correlation, the steps to be taken will be much more complicated

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics

X = df['engine-size'].values.reshape(-1, 1)
y = df['price']

# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
# Predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
# Visualising the Training set results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Price vs Engine-Size (Training set)')
plt.xlabel('Engine-Size')
plt.ylabel('Price')
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, regressor.predict(X_test), color = 'blue')
plt.title('Price vs Engine-Size (Test set)')
plt.xlabel('Engine-Size')
plt.ylabel('Price')
plt.show()
```

```python
mae = metrics.mean_absolute_error(y_test,y_pred)
print("Mean Absolute Error:", mae)
mse = metrics.mean_squared_error (y_test,y_pred)
print("Mean Squared Error:", mse)
rmse = np.sqrt (metrics.mean_squared_error (y_test, y_pred))
print("Root Mean Squared Error:", rmse)
r2 = metrics.explained_variance_score (y_test, y_pred)
print("R-Squared Error:", r2)
```
